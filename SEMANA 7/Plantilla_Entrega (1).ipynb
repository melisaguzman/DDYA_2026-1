## 1) Descripción del problema
En este problema se busca simular el sistema de atención de pacientes en un centro médico usando estructuras de datos. Los pacientes pueden ser de dos tipos: **normales** y **críticos**. Los pacientes normales se atienden en el orden en que llegan, mientras que los pacientes críticos tienen prioridad sobre los normales, aunque entre los pacientes críticos también se respeta el orden de llegada. El sistema recibe varias acciones: agregar un paciente normal (`N nombre`), agregar un paciente crítico (`C nombre`) o atender a un paciente (`A`). Cuando se atiende a un paciente, primero se revisa si hay pacientes críticos; si los hay, se atiende al primero de ellos. Si no hay críticos, se atiende al primer paciente normal. En caso de que no haya pacientes en el sistema, se debe mostrar un mensaje indicando que no hay pacientes. Cada vez que se agrega o se atiende un paciente, se debe mostrar el mensaje correspondiente en la salida.

## 9) Código funcional
Incluye tu implementación completa.



```python
class Cola:

    def __init__(self):
        self.datos = []

    def encolar(self, x):
        self.datos.append(x)

    def desencolar(self):
        if self.datos != []:
            return self.datos.pop(0)

    def frente(self):
        if self.datos != []:
            return self.datos[0]

    def esta_vacia(self):
        return self.datos == []

def atender(criticos, normales):

    if not criticos.esta_vacia():
        p = criticos.desencolar()
        return "Se atiende a " + p

    elif not normales.esta_vacia():
        p = normales.desencolar()
        return "Se atiende a " + p

    else:
        return "no hay pacientes"


def main():

    cola_c = Cola()
    cola_n = Cola()

    n = int(input())

    for i in range(n):

        datos = input().split()

        if datos[0] == "N":

            nombre = datos[1]
            cola_n.encolar(nombre)
            print("Paciente N nuevo :", nombre)

        elif datos[0] == "C":

            nombre = datos[1]
            cola_c.encolar(nombre)
            print("Paciente C nuevo :", nombre)

        elif datos[0] == "A":

            print(atender(cola_c, cola_n))

main()


```
